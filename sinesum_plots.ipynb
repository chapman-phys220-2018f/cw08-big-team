```python
import sinesum
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# t = T/2
# t value can only be adjusted indirectly through T value
T = 2*math.pi

def plot_sum(x,y,n):
    '''
    Convenient way to plot sums'''
    plt.figure(figsize=(10,6))
    plt.plot(x, y)
    plt.plot(x, y, marker = 'o',color='k')
    plt.title("Figure {0}".format(n))
    plt.xlabel('n values')
    plt.ylabel('Sum values')
    plt.show()
```

#### $\alpha$ = 0.01, n = 1,3,5,10,30,100

```python
# a = 0.01
a = 0.01
n = [1,3,5,10,30,100]
y = [sinesum.s(a,n1) for n1 in n]
f = sinesum.f(a*T, T)
plot_sum(n, y, 1)
error_values = abs(np.subtract(np.array(y),f))

print("f value:")
print(f)
print("Sum values: ")
print(y)
print("")
print("Error at each n value |y-f|")
print(error_values)
```

#### $\alpha$ = 0.25, n = 1,3,5,10,30,100

```python
# a = 0.25
a = 0.25
n = [1,3,5,10,30,100]
y = [sinesum.s(a,n1) for n1 in n]
f = sinesum.f(a*T, T)
plot_sum(n, y, 2)
error_values = abs(np.subtract(np.array(y),f))

print("f value:")
print(f)
print("Sum values: ")
print(y)
print("")
print("Error at each n value |y-f|")
print(error_values)
```

#### $\alpha$ = 0.49, n = 1,3,5,10,30,100

```python
# a = 0.49
a = 0.49
n = [1,3,5,10,30,100]
y = [sinesum.s(a,n1) for n1 in n]
f = sinesum.f(a*T, T)
plot_sum(n, y, 3)
error_values = abs(np.subtract(np.array(y),f))

print("f value:")
print(f)
print("Sum values: ")
print(y)
print("")
print("Error at each n value |y-f|")
print(error_values)
```

#### Conclusions
Across all $\alpha$ values, we see that the error (defined as difference of y-values and f) decreases as n increases. This makes sense as with more n values we draw closer to the end behavior of the sequence that approximates the original function. Across $\alpha$ values of 0.01, 0.25, and 0.49, $\alpha=0.25$ is the most accurate by just over a factor of 10. It seems that $\alpha=0.1$ is too low and $\alpha=0.049$ is too high while $\alpha=0.25$ is the Goldilocks value for this problem.

The values for $\alpha=0.1$ and $\alpha=0.049$ are identical. This indicated to me that $\alpha$ values equal to $0.25\pm n$ might be identical. I've supported this assumption with an example below with $n=0.10$, although further exploration is required.

```python
n = [1,3,5,10,30,100]

# a = 0.15
y1 = [sinesum.s(a,n1) for n1 in n]
print(y1, "alpha = 0.15")

# a = 0.35
y2 = [sinesum.s(a,n1) for n1 in n]
print(y2, "alpha = 0.35")
```
